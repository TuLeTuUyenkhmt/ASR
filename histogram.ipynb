```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_hist(img):
    hist = np.zeros(np.max(img)+1)
    h, w = img.shape[0:2]
    for i in range(h):
        for j in range(w):
            hist[round(img[i][j])] += 1
    return hist
```

```python
def print_hist(img):
    img_hist = compute_hist(img)
    x = range(np.min(img), np.max(img)+1)
    plt.bar(x,img_hist)
```

```python
img = np.array([[0,0,1,2],
                [3,0,1,1],
                [3,3,3,2],
                [2,3,1,0]])
plt.imshow(img, cmap = 'gray')
```

```python
print_hist(img)
```

```python
def tg(hist, g):
    return np.sum(hist[0:g+1])
```

```python
img_hist = compute_hist(img)
```

```python
plt.plot([tg(img_hist,i) for i in range(9)])
```

```python
def fg(img, new):
    new_img = img.copy()
    img_hist = compute_hist(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            TB = (img.shape[0]*img.shape[1])/new
            a = tg(img_hist,img[i][j])
            b = np.round(a/TB - 1)
            new_img[i][j] = np.max([0,b])
    return new_img
```

```python
new_img = fg(img, 6)
```

```python
plt.imshow(new_img, cmap = 'gray')
```

```python
print_hist(new_img)
```
